###`rich.print()`
- *The rich.print() function acts as a drop-in replacement for Python's built-in print(), automatically adding syntax highlighting and formatting to data structures.*

```python
from rich import print

# Example: Sales Transaction Data
sale_record = {
    "transaction_id": "TXN-9901",
    "product": "Wireless Mouse",
    "amount": 25.99,
    "tags": ["electronics", "peripherals"],
    "in_stock": True
}

print(sale_record)
```

###`Explanation`:
- *It automatically applies color-coding to strings, numbers, and booleans while pretty-printing dictionaries and lists. This makes debugging complex data structures much faster than standard text output.*

---
###`Table`
- *The Table class allows you to render data in a clean, border-styled grid that is much more readable than a CSV-style printout.*

```python
from rich.table import Table
from rich.console import Console

console = Console()
table = Table(title="Monthly Sales Report")

# Define columns
table.add_column("Product ID", style="cyan", no_wrap=True)
table.add_column("Item Name", style="magenta")
table.add_column("Revenue", justify="right", style="green")

# Add rows
table.add_row("P001", "Gaming Laptop", "$12,400")
table.add_row("P002", "Mechanical Keyboard", "$2,150")
table.add_row("P003", "UltraWide Monitor", "$5,900")

console.print(table)
```

###`Explanation`:
- *Tables are built by adding columns first, then rows, providing a structured way to display summary statistics or dataframe previews. The Console object is required to render the table correctly in the output cell.*

---
###`Progress`
- *The Progress bars are essential for data projects involving long-running loops, such as API scraping or heavy data cleaning.*

```python
import time
from rich.progress import track

products = ["Laptop", "Mouse", "Keyboard", "Monitor", "Headset"]

# Simulating a data cleaning process for sales items
for product in track(products, description="Cleaning Product Data..."):
    time.sleep(0.6)  # Simulating work
    print(f"Finished processing {product}")
console.print("\n[bold green]Pipeline Success![/bold green] :white_check_mark:")
```

###`Explanation`:
- *The track function wraps any iterable to instantly display a live progress bar with percentage and time remaining. It’s the simplest way to keep track of execution status without cluttered log messages.*

###Excercise 2: `Table`
- Task: Build a "Top Performers" table for a sales team.

###Exercise 3: `Progress`
- Task: Simulate a "Bulk Image Upload" process for an e-commerce product catalog.

- Requirements:

- Create a list of 10 dummy image filenames (e.g., img_01.jpg, img_02.jpg).

- Use the track() function to iterate through the list.

- Inside the loop, use time.sleep(0.3) to simulate the upload time.

- Within the same loop, use rich.print() to notify which specific image was "uploaded successfully."

```python
import time
from rich.progress import track
from rich.console import Console
console = Console()
images = [f'img{i:02d}.png' for i in range(1, 11)]
for image in track(images, description='[cyan]Processing Images...[/cyan]'):
    time.sleep(0.3)
    console.print(f"Finished processing [bold magenta]{image}[/bold magenta]")
console.print("\n[bright_red]Pipeline Success![/bright_red] :white_check_mark:")
```
