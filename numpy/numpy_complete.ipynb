## 📌 Chapter 1 — Lists vs NumPy Arrays

Python lists require **loops** to perform operations element by element. NumPy arrays apply operations to **all elements at once**, making it significantly faster and cleaner — especially on large datasets.

### Using a Python Loop (slow approach)

```python
tempreatures = [32.5,45.3,39.2,31.9,28.78]
total = 0
for t in tempreatures:
  total+=t
average = total/len(tempreatures)
print(f"Average tempreature : {average}")
```

### Using NumPy (fast approach)

Same result — no loop needed. NumPy handles it internally with optimized C code.

```python
import numpy as np
tempreatures = np.array([32.5,45.3,39.2,31.9,28.78])
average = np.mean(tempreatures)
print(average)
```

## 📌 Chapter 5 — Indexing & Slicing

### Indexing — Accessing Elements

- **1D**: `arr[index]`
- **2D**: `arr[row, col]`
- **Negative index**: counts from the end

```python
# Indexing in an array
arr = np.array([1,2,3,4,5]) #for 1-d array
print(arr[4])
print(arr[-2])
arr1 = np.array([[1,2,3],[4,5,6]]) #for 2-d array
print(arr1[1,2])
```

### Slicing — Extracting a Range

Syntax: `arr[start : stop : step]`
- Stop index is **exclusive**
- Negative step reverses the array

```python
# slicing of an array
arr = np.array([6,7,8,9,10])
print(arr[2:4]) #using +ve indexes
print(arr[-4:-1]) #using -ve indexes
print(arr[0:4:2]) #start - step - stop
print(arr[::-1])#reverses
```

### Fancy Indexing — Select Multiple Elements at Once

Pass a list of indices to grab multiple specific elements in one go.

```python
#Fancy indexing in an array
#select each element at once
arr = np.array([6,7,8,9,10,12,24,56,77])
print(arr[[-1,-4,0]])
```

### Boolean Masking — Filter by Condition

Returns only the elements that satisfy the condition. You can also combine it with `np.where()` to **replace** values.

```python
#boolean masking to select elements on the basis of condition
arr = np.array([2,14,50,34,32,12,10,43])
print(arr[arr>20])
arr[arr>20]=0
# or use arr = np.where(arr>20,0,arr)
print(arr)
```

### `reshape()` — Change Array Shape

Converts an array to a new shape **without changing the data**. Total elements must remain the same.

```python
#Reshaping an array function
arr = np.array([1,2,3,4,5,5]).reshape(2,3)
brr = np.arange(101,111).reshape(2,5)
crr = np.random.randint(101,111,5)
print(f"{arr}\n{brr}\n{crr}")
```

### `ravel()` vs `flatten()` — Collapse to 1D

| Method | Creates | Modifies Original? |
|--------|---------|-------------------|
| `.ravel()` | View | ✅ Yes — changes reflect back |
| `.flatten()` | Copy | ❌ No — original stays safe |

```python
#.ravel() and .flatten() fucntion uses: multidimen->1-d array...
# ravel -> view(modifies orignal array) while flatten -> copy(orignal array doesnot change)
orignal = np.array([[7,5,9],[23,56,87]])
r = orignal.ravel()
r[0]=921
print(f"{r}\n{orignal}")
f = orignal.flatten()
f[0] = 123
print(f"{f}\n{orignal}")
```

Syntax: `np.insert(arr, index, value, axis=None)`

## 📌 Chapter 7 — Broadcasting & Vectorization

### What is Broadcasting?

Broadcasting lets NumPy apply operations between arrays of **different shapes** — without writing loops. NumPy automatically stretches the smaller array to match the larger one.

### What is Vectorization?

Vectorization means applying operations to **entire arrays at once** instead of element by element. It is the core reason NumPy is faster than plain Python loops.

### Broadcasting Example — Apply Discount to Prices

```python
#BroadCasting using numpy ensuring efficiency and speed...
discount = int(input("Enter Discount : "))
prices = np.array([100,200,300])
final_prices = prices-(prices*discount/100)
print(final_prices)
# same code using for-loops
```

### The Same Task with a Python Loop (for comparison)

```python
#Using forloop is not recommended for large datasets ....
prices = [100,200,300]
discount = int(input("Enter Discount : "))
new = []
for p in prices:
  p = (p-(p*discount/100))
  new.append(p)
print(new)
```

### Scalar Broadcasting

```python
arr = np.array([100,200,300])
result = arr*2
print(result)
```

### 2D × 1D Broadcasting

When a 2D array is multiplied with a 1D array, NumPy broadcasts the 1D array across each row.

```python
arr = np.array([[1,2,3],[5,6,7]])
brr = np.array([10,20,30])
print(arr*brr)
```

### Dimension Mismatch — What Goes Wrong

If shapes are incompatible, NumPy raises a `ValueError`. Use `.reshape()` to fix it.

```python
arr = np.array([[1,2,3],[5,6,7]])
brr = np.array([10,20])
print(arr*brr)
# dimensions must match otherwise numpy will throw ValueError
```

```python
arr = np.array([[1,2,3],[5,6,7]])
brr = np.array([10,20]).reshape(2,1)
print(arr*brr)
# you can use .reshape to fix the above code as per your need
# dimensions must match otherwise numpy will throw ValueError
```

### Vectorization vs List Comprehension

Python list comprehension uses a loop internally. NumPy vectorization uses compiled C — much faster at scale.

```python
l1 = [1,2,3]
l2 = [4,5,6]
result = [x+y for x,y in zip(l1,l2)]
print(result)   # [5, 7, 9]
```

```python
#Vectorication topic:
a1 = np.array([1,2,3])
a2 = np.array([4,5,6])
result = a1+a2
print(result)
```

```python
arr = np.array([1,2,3])
result = arr*3
print(result)
# numpy performs operations at once eliminating the need of for-loops
```

### Feature Standardization using Broadcasting

```python
matrix = np.random.randint(1,21,(5,3))
col_mean = np.mean(matrix,axis=0)
col_std = np.std(matrix,axis=0,keepdims=True)
standardization = (matrix-col_mean)/col_std
print(col_mean)
print(col_std)
print(standardization)
```

### Euclidean Distance — Advanced Broadcasting with `np.newaxis`

```python
a = np.random.randint(1,20,(3,2))
b = np.random.randint(1,20,(4,2))
a = a[:,np.newaxis,:]
# a is now 3-d (3,1,2)
diff = a-b
# a has 3 layers and b has one layer
# a copy each layer 4 times while b copy each layer 3 times and result is [a1-b1,a1-b2,a1-b3,a1-b4]... and so on resulting in a (3,4) matrix
elucidean_distance = np.sqrt(np.sum(diff**2,axis=2))
print(a)
print(b)
print(diff)
print(elucidean_distance)
```

## 📌 Chapter 8 — Handling Missing Values

NumPy uses `np.nan` (Not a Number) to represent missing values. Standard functions like `np.mean()` return `NaN` if any element is `NaN` — use the `nan`-safe versions instead.

| Function | Description |
|----------|-------------|
| `np.isnan(arr)` | Returns `True` where values are `NaN` |
| `np.nan_to_num(arr)` | Replaces `NaN` with `0` (default) |
| `np.nan_to_num(arr, nan=val)` | Replaces `NaN` with a custom value |
| `np.isinf(arr)` | Detects `inf` and `-inf` |
| `np.nanmean(arr)` | Mean — ignores `NaN` |
| `np.nanmax(arr)` | Max — ignores `NaN` |
| `np.nanstd(arr)` | Std deviation — ignores `NaN` |

### Detecting Missing Values with `np.isnan()`

```python
# Handling missing values...
# np.isnan(arr) is used to detect null values in an array true for missing and false for filled values
arr = np.array([1,2,np.nan,5,6,7,np.nan,np.nan])
print(np.isnan(arr))
```

### Replacing NaN with `np.nan_to_num()`

### 🔹 Exercise 2 — Feature Normalization

**Dataset:** 2D feature matrix with two columns (different units/scales).

**Tasks:**
- Apply **min-max normalization** column-wise (scales to 0–1)
- Apply **z-score standardization** column-wise
- Comments explaining when to use each method

```python
import numpy as np

x = np.array([
    [120,  15],
    [150,  18],
    [100,  12],
    [130,  16]
])
max_col = np.max(x,axis=0)
min_col = np.min(x,axis=0)
#finding max and min column wise
'''
Min-max normalization is used to compare values that have known ranges
It transforms each value to 0(low) or 1(high) and 0.5(mid)
Limitations arises if any value is very high like 100000
Z-Score-normalization(standardization) is used to compare values of differ units on large scale
It finishes the differ between large and small values like salary and age respectively
Fixes values at +3 to -3 and if any value is above +3 or below -3 than it is far away from avg
'''
#calculations
max_min_norm = (x-min_col)/(max_col-min_col)
mean_col = np.mean(x,axis=0)
std_col = np.std(x,axis=0)
z_score = (x-mean_col)/std_col
#printing results
print(x)
print(f"\nMax : {max_col}\nMin : {min_col}")
print(f"\nMin-Max-Normalization : \n{max_min_norm}\n\nZ-Score Normalization : \n{z_score}")
```

### 🔹 Exercise 4 — Advanced Indexing & Cleaning

**Dataset:** Weekly revenue array.

**Tasks:**
- Find indices where `revenue > 400`
- Extract those values
- Replace revenues `< 200` with the median
- Sort the cleaned array without modifying the original

```python
import numpy as np

# Weekly revenue data
revenue = np.array([200, 450, 300, 800, 120, 900, 500])
print(f"Original Revenue: {revenue}")

# Finding where revenue is high (above 400)
# 'indices' stores the positions, 'values' stores the actual numbers
indices = np.where(revenue > 400)[0]
values = revenue[revenue > 400]

# Calculating the middle value to use for cleaning the data
median_rev = np.median(revenue)

# If revenue is too low (below 200), replace it with the median to avoid skewing results
cleaned_revenue = np.where(revenue < 200, median_rev, revenue)

# Sorting the cleaned revenue to see the range from lowest to highest
sorted_revenue = np.sort(cleaned_revenue)

# Displaying the findings
def summary():
  print(f"\nPositions where revenue > 400: {indices}")
  print(f"Actual high revenue values: {values}")
  print(f"Calculated Median: {median_rev}")
  print(f"Updated Revenue (low values replaced): {cleaned_revenue}")
  print(f"Final Sorted Revenue: {sorted_revenue}")
summary()
```

### 🔹 Exercise 5 — Time-Series Style Analysis

**Dataset:** 24-hour usage readings.

**Tasks:**
- Reshape to `(24, 1)`
- Compute daily average
- Find hour indices where usage > average
- Compute rolling 3-hour average (no loops)
- Compute percentage change from previous hour

```python

import numpy as np

# Creating the usage array and reshaping it to a 24x1 column vector
usage = np.array([
    30, 28, 27, 25, 26, 29,
    35, 40, 45, 50, 48, 46,
    44, 42, 40, 38, 36, 34,
    32, 31, 30, 29, 28, 27
]).reshape(24,1)

# Calculating the mean value of the entire usage data
daily_average = np.mean(usage)

# Finding indices where usage is higher than the average
# Using [0] to extract row indices from the result tuple
higher_usage = np.where(usage > daily_average)[0]

# Creating three overlapping slices to calculate a 3-hour window
u1, u2, u3 = usage[:-2], usage[1:-1], usage[2:]
rolling_avg = (u1 + u2 + u3) / 3

# Defining current and previous hour slices for comparison
current_features = usage[1:]
previous_features = usage[0:-1]

# Calculating the percentage change (currently using division logic)
percentage_change = (current_features - previous_features) / previous_features * 100

# Printing the results
print(f"Daily_Average : {daily_average}\nHigher_usage_indices : {higher_usage}")
print(f"\nRolling_Average : \n{rolling_avg.flatten()}")
print(f"\nPercentage_change : {percentage_change}")
```
