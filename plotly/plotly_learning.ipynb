###`Plotly` is pre-installed in Google Colab, so you can start importing it immediately.

```python
import plotly.express as px
import polars as pl
```

---
###Loading `Pandas/Polars` DataFrame to apply `Plotly` operations on it

```python
df = pl.DataFrame({
    "category": ["Electronics", "Furniture", "Accessories"],
    "revenue": [5000, 3200, 1500]
})
df_trend = pl.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr"],
    "sales": [200, 450, 300, 600]
})
df_scatter = pl.DataFrame({
    "price": [10, 50, 100, 250, 500, 1000],
    "units_sold": [500, 400, 200, 150, 50, 10]
})
print(df)
print(df_trend)
print(df_scatter)
```

---
###`px.bar`
- *Used to compare categorical data, such as sales across different product categories.*

```python
fig = px.bar(df, x="category", y="revenue", title="Sales by Category",template='plotly_dark')
fig.show()
```

###`Explanation`:
- *px.bar creates a bar chart using the DataFrame and column names for axes. It automatically handles labeling and basic interactivity like hovering.*

---
###`px.line`
- *Perfect for showing trends over time, such as sales growth or price fluctuations.*

```python
fig = px.line(df_trend, x="month", y="sales", title="Monthly Sales Trend",template='plotly_dark')
fig.show()
```

###`Explanation`:
- *px.line connects data points with a line to visualize continuity. It is the go-to choice for time-series analysis in data projects.*

---
###`px.scatter`
- *Ideal for finding correlations between two numerical variables, like checking if higher-priced products sell fewer units.*

```python
fig = px.scatter(df_scatter, x="price", y="units_sold", title="Price vs. Units Sold",template='plotly_dark')
fig.show()
```

###`Explanation`:
- *px.scatter plots individual data points to reveal patterns or outliers. You can add a third dimension by using the size or color parameters.*

---
###`fig.update_layout`
- *This is used to customize the visual style, such as changing titles, axis labels, or the overall theme of the chart.*

```python
fig = px.bar(df, x="category", y="revenue")

# Customizing the appearance
fig.update_layout(
    title="Annual Product Revenue",
    xaxis_title="Product Group",
    yaxis_title="Amount ($)",
    template="plotly_dark"
)
fig.show()
```

###`Explanation`:
- *update_layout allows you to modify non-data elements like themes and labels. It is essential for making charts look professional and "portfolio-ready."*

##🧪Practice Set Of Above Topics:

###Dataset for practice :

```python
import polars as pl
import plotly.express as px

# Main Dataset for Exercises 1, 3, 4, and 5
df_products = pl.DataFrame({
    'Product_ID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Category': ["Electronics", "Accessories", "Electronics", "Furniture", "Accessories", "Furniture", "Electronics", "Accessories"],
    'Base_Price': [1200.50, 25.00, 300.00, 450.00, 15.75, 600.00, 850.00, 45.00],
    'Units_Sold': [15, 120, 45, 8, 200, 12, 25, 150]
})

# Specific Dataset for Exercise 2 (Time Series)
df_time = pl.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Profit": [1200, 1800, 1500, 2200, 2100]
})
```

```python
print(df_products)
print(df_time)
```

---
###Categorical Comparison (`px.bar`)
- Task: Create a bar chart showing the total units sold for each category.

- Logic: You need to group by Category first using Polars.

```python
units_sold_category = df_products.group_by('Category').agg(pl.col('Units_Sold').sum().alias('Total_Units'))
units_sold_category
```

```python
fig = px.bar(units_sold_category,x='Category',y='Total_Units',title='Total_Sales_Per_Category',color='Category',template='plotly_dark')
fig.show()
```

---
###Time-Series Trends (`px.line`)
- Task: Plot the df_time data.

- Goal: Use Month on the X-axis and Profit on the Y-axis.

```python
fig = px.line(df_time,x='Month',y='Profit',title='Profit_Per_Month_Analysis',template='plotly_dark')
fig.show()
```

---
###Correlation Analysis (`px.scatter`)
- Task: Use df_products. Plot Base_Price (X) vs Units_Sold (Y).

- Requirement: Add color="Category" inside the function to differentiate products.

```python
fig = px.scatter(df_products,x='Base_Price',y='Units_Sold',title='Units_Sold vs Base_Price',color='Category',size='Units_Sold')
fig.show()
```

---
###The Professional Touch (`fig.update_layout`)
- Task: Take your code from Task 3.

- Requirement: Add .update_layout() to change the labels and set template="plotly_dark".

```python
fig.update_layout(xaxis_title='Products_Price',yaxis_title='Sold_Value',title='Products_Price vs Sold_Value',template='plotly_dark')
fig.show()
```

---
###The Integration (`Polars Lazy + Plotly`)
- Task: Create a lazy chain: df_products.lazy() -> group_by -> sum(Gross_Revenue) -> collect().

- Goal: Pass that final result directly into px.bar.

```python
df_lazy = (
    df_products.lazy()
    .group_by('Category').agg(pl.col('Units_Sold').sum().alias('Total_Units')).collect()
)
fig = px.bar(df_lazy,x='Category',y='Total_Units',title='Total_Units per Category',color='Category',template='plotly_dark')
fig.show()
```
